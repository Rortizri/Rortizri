```python
#1. Generate 200 points from the unit square: for example, draw the x1 coordinates from a U[0, 1] distribution and the x2 coordinates from a U[0, 1] .

n = 200

x1 = np.random.uniform(0,1,n)
print(x1)

```

```python
x2 = np.random.uniform(0,1,n)
print(x2)
```

```python
# 2. Assign the label “blue” and “orange” to each of the 200 observations according to the following rule:• if 2 × x1 + 0.25 − x2 > 0 and −2 × x1 + 2 − x2 > 0, then the label should be “orange”• otherwise, the label should be “blue”.

label = []
for i in np.arange(n):
 if (2*x1[i]-x2[i]+ .25>0) and (-2*x1[i]+2-x2[i]>0):
    label.append("orange")
 else:    
    label.append("blue")
```

```python
# 3. Now, add a noise to each point. The noise should be randomly drawn from a U[−.1, .1] (this way, the 200 generated points should not be perfectly separable).

n = 200 
noise = np.random.uniform(-1,1,n)

nx1 = x1 + noise
print(nx1)
```
